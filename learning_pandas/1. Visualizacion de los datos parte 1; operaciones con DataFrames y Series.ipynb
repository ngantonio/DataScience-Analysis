```python
#Cargamos un dataset de peliculas
movieusers = pd.read_table('http://bit.ly/movieusers')
```

```python
# Si solo imprimimos movieusers, imprimira una sola columna ya que no reconoce wl separador |
movieusers.head()
```

```python
# Le señalamos Pandas que se usará este separador : | e imprimimos nuevamente:
movieusers = pd.read_table('http://bit.ly/movieusers', sep='|')
movieusers.head()
```

```python
#Podemos eliminar la cabecera de los datos:
movieusers = pd.read_table('http://bit.ly/movieusers', sep='|', header= None)
movieusers.head()
```

```python
# Podemos crear una nueva cabecera para nuestros datos, y modificar la que tenia por defecto, creando una lista:

#Creamos la lista
col_names = ['ID', 'Age', 'Gender', 'Occupation','Revenue']

# le decimos que los nombres de las columnas estaran en una variable llamada "col_names"
movieusers = pd.read_table('http://bit.ly/movieusers', sep='|', header= None, names=col_names)

movieusers.head()
```

## Operaciones con dataframes y Series:

### Se pueden hacer operaciones basicas de tipos con las series, siempre y cuando sus conversiones sean compatibles

```python
# Intentando concatenar numeros con strings: ERROR
# movieusers.Age + movieusers.Gender

# Intentando concatenar string con strings:
movieusers.Occupation + " " + movieusers.Gender

# Ahora bien, podemos asignar el resultado de una opedracion entre conjuntos, a una nueva columna:
movieusers['concatenate'] = movieusers.Occupation + " " + movieusers.Gender
movieusers.head()
```

### Nombres y operaciones con columnas:

```python
#Muestra los nombres de las columnas:
data.columns
```

```python
#Puedo reemplazar el nombre de una columna:

#Sin alterar el data frame:
data.rename(columns={'star_rating':'starRating'})

#Alterando el data frame:
data.rename(columns={'star_rating':'starRating'}, inplace=True)
data.head()
```

```python
#Podemos reemplazar el contenido de las columnas:

data_cols = ['StarRating','Title','ContentRating','Genre','Duration','Actors']
data.columns = data_cols
data.head()
```

```python
#Podemos reemplazar substrings dentro de los nombres de columnas:
#la siguiente linea, quita los guiones bajos de los nombres:

data.columns = data.columns.str.replace('_','')
data.head()
```
