## FILTRADO MEDIANTE CRITERIOS MULTIPLES



```python
import pandas as pd
```

```python
data = pd.read_csv('http://bit.ly/imdbratings')
data.head()
```

### Todas las pelis del genero Crime, y que su rating sea R:

```python
data[(data.content_rating == 'R') & (data.genre == "Crime")].head()

# ¿Como hace esto?, pues como el ejemplo de la lista de booleanos, recorre todas las filas
# y crea una lista de n filas, (n es el tamaño de las filas del dataframe), y coloca True or False 
# segun se cumpla o no la condicion, este es el resultado de esa lista:

data.content_rating == 'R'

# luego, para ejecutar las consultas, solo selecciona las filas del dataframe cuya posicion (index) 
# coincida con las filas de la lista de booleanos que esten en True.

data[(data.content_rating == 'R') & (data.genre == "Crime")].head()
```

```python
# Podemos hacer filtrado multiple:
# esta seria la forma simple de conocer todas las pelis del genero Drama, Crime and Action

data[(data.genre == "Crime") | (data.genre == "Action") | (data.genre == "Drama")].head()
```

```python
# Pero pandas nos ofrece una funcion que nos permite agrupar consultas para una columna de datos:
# Si imprimimos esto, el resultado será una lista de booleanos, equivalente a la linea anterior
data.genre.isin(['Crime', 'Action', 'Drama'])

#de esta forma
data[data.genre.isin(['Crime', 'Action', 'Drama'])].head()
```

```python
# Podemos incluso definir el tipo de dato que tendran algunas columnas, 
# en el mismo momento en el cual cargamos el dataframe, proporcionando un diccionario:

data = pd.read_csv('http://bit.ly/imdbratings', dtype = {'duration':float})
data.dtypes
```

```python
#cargaremos el data set de un restaurante:
data = pd.read_table('http://bit.ly/chiporders')
data.head()
```

```python
# si mostramos los tipos de datos, nos damos cuenta que el precio es object, 
# ademas de tener el simbolo de $, no es un dato numerico puro:
data.dtypes
```

```python
#primero deberiamos de quitar ese simbolo de $ (limpieza de datos):
data.item_price = data.item_price.str.replace('$', '')
data.head()
```

```python
#listo, ahora podemos convertir la serie en flotante:
data.item_price = data.item_price.astype(float)

#verificamos
data.dtypes
```

```python
#queremos la media de todos los precios:
data.item_price.mean()
```

# Dando significado a los datos:

### Muchas veces debemos convertir datos, para darles un significado, los algoritmos de machine learning, solo entienden de números, a los cuales nosotros les proporcionamos un significado.

#### Tambien para obtener datos completos como la moda, media o varianza:


```python
# Queremos un dataset que contenga todos los platos que tengan la palabra "salsa" en sus titulos
# para enviarlo a un modelo de machine laerning:

#hacemos una copia del data set
new_salsa = data
new_salsa.head()

#estos son todos los platillos que contienen salsa, pero el resultado es presentado en String, "True" or "False"
new_salsa.item_name.str.contains("Salsa")

#lo necesitamos en digito: 1 y 0, asi que lo convertimos:
new_salsa.item_name.str.contains("Salsa").astype(int).head()
```

```python
#ahora acualizamos el data set con esos 1' s y 0´s:
new_salsa.item_name = new_salsa.item_name.str.contains("Salsa").astype(int)
new_salsa.head()
```

```python
#Ahora, el nuevo data set se divide en 2, los platos que contienen salsa y los que no:
new_salsa.dtypes
```

```python
#si quiero los que tienen salsa:
new_salsa[new_salsa.item_name == 1].head()
```
