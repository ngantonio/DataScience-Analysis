```python
import pandas as pd
```

```python
#Cargamos el dataset de peliculas
data = pd.read_csv('http://bit.ly/imdbratings')
data.head()
```

## Las Agrupaciones:

### Podemos agrupar todo el dataset, por los datos de alguna columna en especifico para encontrar datos estadisticos:

```python
# Agrupando por genero:

#El promedio de la duracion de las peliculas de todos los generos
data.groupby('genre').duration.mean()
```

```python
# La menor duracion de una pelicula por genero:
data.groupby('genre').duration.min()
```

### Tabla de estadisticos:

```python
# Podemos seguir imprimiendo valores asi, o podemos decirle a pandas que nos imprima 
#una tabla de estadisticos de la duracion de las peliculas por genero, con la funcion agg:

#Agrupamos el dataset por genre, y aplicamos todas estas funciones a la columna duration
data.groupby('genre').duration.agg(['count','min','max','median','mean'])

#imprime por genero, el total de peliculas, la de menor, media y menor duracion y el promedio:
```

```python
# Estadisticos en general agrupados por genero
# para cada columna de la tabla (duracion y rating) aplica las funciones:

data.groupby('genre').agg(['count','min','max','median','mean'])
```

### Funciones Disponibles para Pandas.Series

```python
# estadisticos mas reelevantes de una columna:

# freq  es el mayor numero de incidencias
data.genre.describe()
```

```python
# Cuantos valores unicos existen para el genero de la pelicula
data.genre.value_counts()
```

```python
# valor anterior expresado en porcentaje: 
data.genre.value_counts(normalize=True)
```

```python
#regresa un Array con los valores unicos de esa columna
data.genre.unique()
```

```python
# regresa la cantidad de valores unicos de una columna:
data.genre.nunique()
```

## Crea una tabla comparativa entre cada item unico de una columna, versus cada item unico de otra columna


```python
# para este ejemplo, se muestra (genero versus rating)

# la cantidad de peliculas clasificadas por su contenido, para cada genero:
# por ejemplo, de las peliculas de animacion, solo 25 de ellas son de tipo PG

pd.crosstab(data.genre, data.content_rating)
```
